Load excel data and plot average temperature by Year

```python
%matplotlib inline

import pandas as pd

#Â Can read from an URL!
amt = pd.read_excel('http://sdwebx.worldbank.org/climateportal/DownloadData/tas_1991_2015.xls',
                    # This is to avoid strange characters in the file such as "\t" in Year
                   names=['tas', 'year', 'month', 'country', 'ISO3', 'ISO2'])
```

##Â Get some description

```python
##Â We are interested in the temperature average per year
```

```python
tas = amt.groupby('year')['tas']
tas.describe()
```

```python
tas.mean().plot()
```
