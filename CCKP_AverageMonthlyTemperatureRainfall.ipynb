## Download dataset from the Climate Change Knowledge portal
1.	Visit http://sdwebx.worldbank.org/climateportal/index.cfm
2.	Click on the area, then country of your interest on the interactive map
3.	Click the ‘Click to download historical data’ link
4.	Select ‘Temperature’, the country and the time period you are interested in
5.	Click ‘Download Data To Excel’, and save the file on your computer as temperatures.xls
6.	Select ‘Rainfall’, the country and the time period you are interested in
7.	Click ‘Download Data To Excel’, and save the file on your computer as rainfall.xls



## Read raw data from file

Upload temperatures.xls
Upload rainfall.xls

```r
library(gdata)
raw_temp <- read.xls("temperatures.xls")
raw_rain <- read.xls("rainfall.xls")
```

## Peek data

```r
head(raw_temp)
head(raw_rain)
```

## Aggregate datasets and calculate the mean

```r
datasets_temp = aggregate(raw_temp[, 1:2], list(raw_temp$X.Year), mean)
datasets_rain = aggregate(raw_rain[, 1:2], list(raw_rain$X.Year), mean)
```

## Check datasets

```r
print(datasets_temp)
print(datasets_rain)
```

## Drop duplicate column (X.Year) from datasets and plot results

```r
plot (datasets_temp[-3], type="l", col="blue", main="Average Monthly Temperature/Rainfall per year", xlab="year", ylab="Temperature/Rainfall")
par(new=TRUE)
plot(datasets_rain[-3], type="l", col="green", xlab="year", ylab="")
```
